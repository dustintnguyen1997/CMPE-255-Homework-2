```python
import pandas as pd
df = pd.read_csv('coin_Dogecoin.csv')
```

```python
print(df.columns)
```

```python
# import pycaret anomaly and init setup
from pycaret.anomaly import *
s = setup(df, session_id = 123)
```

```python
# train iforest model
iforest = create_model('iforest')
iforest
```

```python
# to check all the available models
models()
```

```python
iforest_anomalies = assign_model(iforest)
iforest_anomalies
```

```python
# tsne plot anomalies
plot_model(iforest, plot = 'tsne')
```

```python
# predict on test set
iforest_pred = predict_model(iforest, data=df)
iforest_pred
```

```python
# save pipeline
save_model(iforest, 'iforest_pipeline')
```

```python
# load pipeline
loaded_iforest_pipeline = load_model('iforest_pipeline')
loaded_iforest_pipeline
```
